```python
df = df.dropna()
df
```

```python
df["price"] = df["price"].replace({'\€':''}, regex = True)
df["price"] = df["price"].replace({' ':''}, regex = True)
df
```

```python
df["price"] = df["price"].replace({',':'.'}, regex = True)
df
```

```python
df["rating"] = df["rating"].replace({'von 5 Sternen':'/ 5'}, regex = True)
df
```

```python
df["rating"] = df["rating"].replace({'out of 5 stars':'/ 5'}, regex = True)
df.head(15)
```

```python
df["rating"] = df["rating"].replace({',':'.'}, regex = True)
df
```
