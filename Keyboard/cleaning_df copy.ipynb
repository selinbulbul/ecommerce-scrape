```python
import pandas as pd
```

```python
df = pd.read_json("search_output.jsonl")
```

# PRICE

```python
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
df["price"] = df["price"].replace({'\€':''}, regex = True)
df.head(20)
```

```python
df["price"] = df["price"].replace({',':'.'}, regex = True)
df.head(15)
```

```python
df["rating"] = df["rating"].replace({'von 5 Sternen':'/ 5'}, regex = True)
df.head(15)
```

```python
df["rating"] = df["rating"].replace({'out of 5 stars':'/ 5'}, regex = True)
df.head(15)
```

```python
df["rating"] = df["rating"].replace({',':'.'}, regex = True)
df.head(15)
```

```python
### additional split function  df.price[0].split('$')[1] , df.rating[0].split(' ')[0] , https://pandas.pydata.org/docs/reference/api/pandas.Series.str.split.html
```

```python
df.loc[:, 'product'] = 'DSLR_Cameras'

df.head(15)
```
