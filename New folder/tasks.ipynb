### 9th Task- Quarterly revenue ( How about creating a plotly express animation)

#### What is the overall sales trend?

```python
df.columns
```

```python
results = df.groupby("invoice_month").sum()
months = [month for month, df in df.groupby("invoice_month")]
plt.figure(figsize=(15,5))
plt.plot(months, results["total_sale"], color ="#b80045")
plt.xticks (months, rotation = "vertical", size =8)
plt.ylabel ("Sales in €")
plt.xlabel("Month")
plt.show();
```

#### What is the overall MoM, QoQ and YoY growth?

```python
df.columns
```

```python
df8_MoM_1= df.groupby(["invoice_month", "month"])["total_sale"].sum().reset_index()
df8_MoM_1
```

```python
df8_MoM_1['Last_Month'] = np.roll(df8_MoM_1['total_sale'],1)
```

```python
df8_MoM_1
```

```python
df8_MoM_1['Growth'] = (df8_MoM_1['total_sale']/df8_MoM_1['Last_Month'])-1
```

```python
p = sb.catplot(x = 'invoice_month', y = 'Growth', kind="bar", data = df8_MoM_1, ax=ax, color ="#b80045")
_, xlabels = plt.xticks()
p.set_xticklabels(xlabels, size=10, rotation = "vertical")
```

```python
results = df8_MoM_1.groupby("invoice_month").sum()
months = [month for month, df8_MoM_1 in df8_MoM_1.groupby("invoice_month")]
plt.figure(figsize=(15,5))
plt.bar(months, results["Growth"], color ="#b80045")
plt.xticks (months, rotation = "vertical", size =8)
plt.ylabel ("% Growth")
plt.xlabel("Month")
plt.show();
```
